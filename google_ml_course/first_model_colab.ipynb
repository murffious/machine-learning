```
import tensorflow as tf
import numpy as np
from tensorflow import keras

# Define and compile the neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
# add appropriate loss and optimizer functions for different scenarios
model.compile(optimizer='sgd', loss='mean_squared_error')

# Provide the data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```
# Train the nueral network
model.fit(xs, ys, epochs=500)
# note if loss is small model is doing a great job of inferring the relationship between the numbers
# eg Epoch 111/500 loss drops below 0.0
```

```
# Use the model - figure out the Y for a previously unknown X.
# model that has been trained to learn the relationship between X and Y. example, if X is 10, what do you think Y will be?
print(model.predict([10.0]))
```
